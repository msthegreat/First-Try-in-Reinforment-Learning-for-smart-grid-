```python
from battery_env import BatteryEnv
from agent import Agent
from tabulate import tabulate

if __name__ == "__main__":
    num_episodes = 10
    num_steps_per_episode = 24

    best_rewards_per_episode = []

    for episode in range(num_episodes):
        env = BatteryEnv()
        agent = Agent()

        current_obs = env.reset()
        print(f"Episode {episode + 1}/{num_episodes}")

        episode_rewards = []  # List to store rewards for each step in the episode

        for step in range(num_steps_per_episode):
            print(f"Step {step + 1}/{num_steps_per_episode}")
            current_obs = agent.step(current_obs, env)
            episode_rewards.append(agent.best_reward_per_episode[-1])  # Store the reward for this step

        best_reward_per_episode = max(episode_rewards)  # Get the best reward for this episode
        best_rewards_per_episode.append(best_reward_per_episode)
        print(f"Episode {episode + 1} - Best Reward: {best_reward_per_episode}\n")

    # Print the best rewards in a table
    table_data = [(f"Episode {i + 1}", best_reward) for i, best_reward in enumerate(best_rewards_per_episode)]
    table_headers = ["Episode", "Best Reward"]
    print(tabulate(table_data, headers=table_headers, tablefmt="grid"))
```

```python
import numpy as np
import pandas as pd
from battery_env import BatteryEnv

# Function to run an episode using a random policy with 24 steps
def run_random_episode(env):
    episode_data = []
    state = env.reset()
    total_reward = 0
    done = False
    step = 1
    while not done and step <= 24:  # Limiting the episode to 24 steps
        action = np.random.randint(3)  # Choose a random action (0, 1, or 2)
        next_state, reward, done, _ = env.step(action)
        total_reward += reward
        episode_data.append([step, action, next_state[0], next_state[1], next_state[2], env.total_power, reward, total_reward])
        step += 1
    return pd.DataFrame(episode_data, columns=["Step", "Action", "SOC", "Pb", "Pg", "Total Power", "Reward", "Cumulative Reward"])

# Create an instance of the BatteryEnv with specified power range
power_range = (800, 1200)  # Set the power range as needed
env = BatteryEnv(power_range)

# Run multiple episodes with 24 steps using a random policy and store the results
num_episodes = 10
episode_data_list = []
for episode in range(num_episodes):
    episode_data = run_random_episode(env)
    episode_data_list.append(episode_data)

# Concatenate the data from all episodes
result_df = pd.concat(episode_data_list, keys=range(1, num_episodes + 1))

# Display the results in a tabular format
result_df
```
